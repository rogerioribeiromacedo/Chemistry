<h1 style="color: #2e6c80;">1. Importando Bibliotecas e configurações</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.float_format = '{:.15f}'.format

# mostrando todas as colunas do dataframe
pd.set_option('display.max_columns', None)
# mostrando x linhas
pd.set_option('display.max_rows', 25)
```

<h1 style="color: #2e6c80;">2. Dados para titulação</h1>

```python
#
# Dados iniciais
#
c_base = 0.1000 # concentração da base (NaOH)
c_acido = 0.1000 # concentração do ácidos
v_acido = 0.05 # 50 ml - volume do ácido (em L)
v_base_adicionada = 0.00 # controla a qtde de base adicionada (em L)
```

<h1 style="color: #2e6c80;">3. Lista de ácidos fracos</h1>
A lista contempla 10 ácidos fracos. Para cada um o respectivo valor de Ka e pKa</br></br>

```python
#
# Dataframe com lista de acidos fracos
#
# Ácido acético (CH3COOH).......: Ka = 1,75 x 10^-5  - pKa =  4.756
# Ácido benzoico (C6H5CO2H).....: Ka = 6,25 x 10^-5  - pka =  4.204
# Ácido ciânico (HCNO)..........: Ka = 3.50 x 10^-4  - pKa =  3.460 
# Ácido fórmico (CH2O2).........: Ka = 1.80 x 10^-4  - pKa =  3.750
# Ácido hidrazoico (HN3)........: Ka = 2.50 x 10^-5  - pKa =  4.602
# Ácido hidrociânico (HCN)......: Ka = 6.20 x 10^-10 - pKa =  9.207
# Ácido fluorídrico (HF)........: Ka = 6.30 × 10^-4  - pKa =  3.200
# Ácido iodoacético (CH2ICO2H)..: Ka = 6.60 × 10^-4  - pKa =  3.180
# Ácido nitroso (HNO2)..........: Ka = 5.60 × 10^-4  - pKa =  3.252
# Ácido hipoiodoso (HIO)........: Ka = 3.20 x 10^-11 - pKa = 10.495
#
acidos_fracos = {"Acido" : ["Ácido acético", "Ácido benzoico", "Ácido ciânico", "Ácido fórmico", "Ácido hidrazoico",
                            "Ácido hidrociâncio", "Ácido fluorídrico", "Ácido iodoacético", "Ácido nitroso", "Ácido hipoiodoso"],
                "Formula" : ["CH3COOH", "C6H5CO2H", "HCNO", "CH2O2", "HN3", "HCN", "HF", "CH2ICO2H", "HNO2", "HIO"],
                 "Ka" : [1.75*10**-5,
                         6.25*10**-5,
                         3.50*10**-4,
                         1.80*10**-4,
                         2.50*10**-5, 6.20*10**-10, 6.30*10**-4, 6.60*10**-4, 5.60*10**-4, 3.20*10**-11
                        ],
                "pKa" : [4.756, 4.204, 3.460, 3.750, 4.602, 9.207, 3.200, 3.180, 3.252, 10.495]}
df_acidos_fracos = pd.DataFrame(acidos_fracos)
df_acidos_fracos
#df = df_acidos_fracos.loc[(df_acidos_fracos["Acidos"] == "CH3COOH")]
#df
```

```python
# Ordenando os dados pela coluna do Ka
df_acidos_fracos.sort_values("Ka", inplace=True)
```

<h1 style="color: #2e6c80;">4. Gerando Dados</h1>

```python
# df = df_acidos_fracos.loc[(df_acidos_fracos["Acidos"] == "CH3COOH") | (df_acidos_fracos["Acidos"] == "H3PO43")]
resultados = []
for index, row in df_acidos_fracos.iterrows():
    # Indentificando o ácido
    acido = row['Acido']
    ka = row['Ka']
    pKa = row['pKa']
    formula = row['Formula']

    # Variável que contempla o pH a ser utilizado
    pH = 0
    concentracao_H30 = 0
    # Considerando que uma gota teria aproximadamente 0,05 mL
    for volume_da_base_ad in np.arange(0, 0.1205, 0.00005): # volume em Litros
        if volume_da_base_ad == 0:
            # No ponto inicial (sem adição de NaOH)
            concentracao_H3O = np.sqrt(ka * c_acido)
            pH = -np.log10(concentracao_H3O)
            
            # adicionando resultado
            resultados.append([acido, np.around(pH, 5), volume_da_base_ad])
        else:
            if volume_da_base_ad < 0.05:
                # formação de uma solução tampão
                # antes do ponto de equivalência
                concentracao_sal = (c_base * volume_da_base_ad) / (v_acido + volume_da_base_ad)
                concentracao_acido = ((c_acido * v_acido) - (c_base * volume_da_base_ad)) / (v_acido + volume_da_base_ad)
                log_sal_acido = np.log10(concentracao_sal / concentracao_acido)                
                pH = pKa + log_sal_acido

                # adicionando resultado
                resultados.append([acido, np.around(pH, 5), volume_da_base_ad])
            else:
                if volume_da_base_ad == 0.05:
                    # Estamos no ponto de equivalência (PE)
                    # o pH é determinado pela hidrólise do sal formado
                    concentracao_sal = (c_base * volume_da_base_ad) / (v_acido + volume_da_base_ad)
                    kb =  (1.00 * 10**-14) / (ka)
                    concentracao_OH = np.sqrt(kb * concentracao_sal)
                    pOH = -np.log10(concentracao_OH)
                    pH = 14.0000 - pOH

                    # adicionando resultado
                    resultados.append([acido, np.around(pH, 5), volume_da_base_ad])
                else:
                    # Excesso de NaOH (inibe a hidrólise do sal)
                    concentracao_OH = ( (c_base * volume_da_base_ad) - (c_acido * v_acido) ) / (v_acido + volume_da_base_ad)
                    pOH = -np.log10(concentracao_OH)
                    pH = 14.0000 - pOH

                    # adicionando resultado
                    resultados.append([acido, np.around(pH, 5), volume_da_base_ad])

# Dataframe de resultados
df_resultados = pd.DataFrame(resultados, columns=["Acido", "pH", "Volume"])
```

<h2 style="color: #2e6c80;">4.1 Valores de pH mínimo e máximo para cada ácido</h2>
